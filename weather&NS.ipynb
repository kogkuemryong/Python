```python
# 날짜 데이터 타입으로 변경
weather['날짜'] = pd.to_datetime(weather['날짜'])
weather.dtypes
```

```python
weather['year'] = weather['날짜'].dt.year
weather['month'] = weather['날짜'].dt.month
weather['dayofweek'] = weather['날짜'].dt.isocalendar().week
weather['date'] = weather['날짜'].dt.day
```
