```python
# 정리된 데이터 읽어오기
mydata = pd.read_csv('mydata.csv', encoding='cp949')
```

```python
mydata
```

```python
mydata.corr()
```

```python
mydata.drop(['남성'], axis=1, inplace=True)
```

```python
mydata.drop(['여성'], axis=1, inplace=True)
```

```python
mydata['sex'] = mydata['상품명']
```

```python
for i in range(len(mydata)):
    if set(mydata['sex'][i].split()) & {"남성"} :
        mydata['sex'][i] = "남성"
    elif set(mydata['sex'][i].split()) & {"여성"} :
        mydata['sex'][i] = "여성"
    else :
        mydata['sex'][i] = "무관"
```

```python
mydata
```

```python
mydata['남성'] = np.where(mydata['sex']=='남성', 1 ,0)
```

```python
mydata['여성'] = np.where(mydata['sex']=='여성', 1 ,0)
```

```python
mydata['무관'] = np.where(mydata['sex']=='무관', 1 ,0)
```

```python
mydata
```

```python
mydata.drop(['sex'],axis=1, inplace=True)
```
