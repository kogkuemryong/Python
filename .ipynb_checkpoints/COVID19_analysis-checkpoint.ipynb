```python
# 데이터 타입을 변경해서 날짜형태로 변환
# 판다스의 to._datetime 을 사용해서 날짜 타입으로 변경
# 연도가 없기 때문에 2020년 날짜에 추가 하고 "-" 문자로 날짜를 연결
df['확진일자'] = pd.to_datetime("2020"+"-"+df['확진일'].str.replace(".","-"))
df['확진일자']
```

##### 3월에 가장 큰 폭을 보이고 4월부터 5월 초까지 거의 줄어들다가 6월부터 다시 상승하였다.

```python
# 선그래프에 값 표시하기
# 일자별 확지자수를 선그래프로 그림
# 연도는 모두 2020년이기 때문에 월만 표기하도록 슬라이싱을 사용해 '월일' 컬럼을 만듬

# astype() 메소드를 이용한 문자열 칼럼의 숫자형 변환
# map은 리스트의 요소를 지정된 함수로 처리해주는 함수(map은 원본 리스트를 변경하지 않고 새 리스트를 생성).
```

```python
df['월일'] = df['확진일자'].astype(str).map(lambda x : x[-5:]) # 문자열로 바꾸고 뒤에서 부터 출력하여 년(2020) 제외
day_count = df['월일'].value_counts().sort_index()
g = day_count.plot(figsize = (15,4), rot = 30)

for i in range(len(day_count)):
    case_count = day_count.iloc[i] # iloc - 위치 정수를 기반으로 인덱싱을 함
    if case_count > 20:
        g.text(x = i-0.3, y = day_count.iloc[i], s = day_count.iloc[i])
```

```python
# 모든 날짜를 행에 만들어주기 
# 첫 확진일 부터 마지막 확진일
# 데이터프레임의 첫번째 날짜는 first_day 에 마지막 날짜는 last_day

first_day = df.iloc[-1]['확진일자']
last_day = df.iloc[0]['확진일자']
first_day, last_day 
```

```python
# pd.date_range 를 통해 시작 날짜부터 끝나는 날짜까지
# DatetimeIndex 를 만들고 days 라는 변수에 저장

days = pd.date_range(first_day, last_day)
days
```

```python
# days 변수의 값으로 '연월일' 이라는 컬럼이름을 갖는 데이터 프레임 만듬
days = pd.DataFrame({'확진일자': days})
days.head()
```

```python
# 확진일자별로 빈도수 구하기 
daily_case = df['확진일자'].value_counts()
daily_case
```

```python
# 확진일자별로 빈도수 구한 내용을 데이터프레임으로 변환하기 
daily_case = daily_case.to_frame()
daily_case.head()
```

```python
# value_counts 결과의 컬럼명을 '확진수' 로 바꾸기 
daily_case.columns = ['확진수']
daily_case.head()
```

```python
# merge 를 통해 전체 확진일자 만들기
# 확진자가 없는 날도 일자에 표현되도록 전체 일자와 확진 데이터를 merge 로 합쳐줍니다. 
all_day = days.merge(daily_case, left_on = "확진일자", right_on = daily_case.index, how = 'left')
all_day = all_day.rename(columns={'연번':'확진수'})
all_day
```

##### 확진자가 없는 날 - NaN

```python
# 누적 확진자 수 구하기 
# 확진수를 fillna를 통해 결측치를 0으로 채워주고 누적해서 더함
all_day['누적확진'] = all_day['확진수'].fillna(0).cumsum()
all_day
```

```python
# 연도를 제외하고 월 - 일 컬럼 만들기 
all_day['일자'] = all_day['확진일자'].astype(str).map(lambda x: x[-5:])
all_day['일자'].head()
```

```python
# "확진수", "누적확진" 컬럼을 갖는 데이터프레임을 만듬
cum_day = all_day.set_index('일자')
cum_day = cum_day[['확진수','누적확진']]
cum_day
```

```python
# 데이터프레임 그리기 확진수와 누적확지을 선그래프로
cum_day.plot(figsize=(15,4))
```

##### 누적 확진자수가 너무 커서 확진수가  잘 보이지 않는다. 

```python
# 시리즈로 2개의 그리프 그리기
cum_day['누적확진'].plot(label ='누적확진', figsize=(15,4))
cum_day['확진수'].plot(label ='확진수', figsize=(15,4))
```

```python
# 누적확진수와 확지수와 차이가 크면 제대로 보이지 않기 때무에 확진수만 시각화

# 확진수 시각화
cum_day['확진수'].plot(figsize=(15,4))
```

```python
# 누적확진수 
cum_day['누적확진'].plot(figsize=(15,4))
```

```python
# 로그스케일
# 차이가 너무 커서 그래프가 자세히 보이지 않을때 로그스케일로 표현하면 차이가 큰 값의 스케일을 조정
np.log(cum_day['누적확진']).plot(figsize = (15,4))
np.log(cum_day['확진수']).plot(figsize = (15,4))
```

```python
# 확진원과 요일 구하기
all_day['확진월'] = all_day['확진일자'].dt.month
all_day['확진요일'] = all_day['확진일자'].dt.dayofweek
all_day.head()
```

```python
# 월별, 요일별 확진수를 groupdy로 구하기 
all_day_week = all_day.groupby(
    ['확진월', '확진요일'])['확진수'].sum().unstack().astype(int)

all_day_week
```

```python
# 숫자로 표현된 요일을 문자로 바꾸기 위한 split 을 통해 문자를 리스트로 변경
dayofweek = '월 화 수 목 금 토 일'
dayofweek = dayofweek.split()
dayofweek
```

```python
# 컬럼의 이름을 한글요일명으로 변경
all_day_week.columns = dayofweek
all_day_week 
```

```python
# style.background_gradient 로 색상을 표현
all_day_week.style.background_gradient(cmap='Blues')
```

```python
# 거주지별 확진자의 빈도수 
gu_count = df['거주지'].value_counts()
gu_count.head()
```

```python
# 구별 확진자의 수 시각화
gu_count.sort_values().plot.barh(figsize = (10,12), grid=True)
```

```python
# 서울에서 확진판정을 받은 데이터이기 때문에 거주지가 서울이 아닐 수 있다. 
# 거주지 별로 서울시에 해당되는 데이터만 다로 가져옴 

gu = ['강남구', '강동구', '강북구', '강서구', '관악구','광진구','구로구', '금천구',
         '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구'
       '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구','중랑구'] 
gu[:5]
```

```python
# 거주지가 서울이 아닌 지역을 따로 추출
set(gu_count.index) - set(gu)
```

```python
# 구를 전처리 하기 쉽게 컬럼으로 변환하기 위헤 reset_index 로 변환
df_gu = gu_count.reset_index()
df_gu.columns = ['구', '환자수']
df_gu.head()
```

```python
# 서울시에서 확진 받은 사람 중 서울 vs 타지역을 비교해 보기 
# '지역' 이라는 새로운 커럶을 만들어 서울지역이 아니라면 '티지역' 이라는 값을 넣음
#.loc[행인덱스]
#.loc[행, 열]
#.loc[조건, 열]
df.loc[df['거주지'].isin(gu),'지역'] = df['거주지']
df.loc[- df['거주지'].isin(gu),'지역'] = '타지역'
df.head()
```

```python
# ' 지역 ' 컬럼으로 확진자 빈도수 구하기
df_gu_etc = df['지역'].value_counts()
df_gu_etc
```
